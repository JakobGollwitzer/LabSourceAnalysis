Jupyter Notebook for processing raw data from the Rigaku Lab X-ray Diffraction Source. Here theta 2 theta curves are extracted from the raw data files.

```python
### import modules
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
### import data  for 002 peak for RT
path = '/data2/jg2364/XRD_Data/YES187/9_28_00L/9_28_00L_RT/'
filename = "th2th.ras"
f = open(path+filename, 'rb')
thelines = [p for p in f]
#### lines in the data for the peak
lines = thelines[380:1430]


### put the data in an array and save
table = np.zeros((len(lines), 2))

for i in range(0, len(lines), 1):
    table[i][0] = lines[i].strip().split()[0]
    table[i][1] = lines[i].strip().split()[1]


rt_4_y = np.zeros(int(len(table[:, 1])))
rt_4_x = np.zeros(int(len(table[:, 1])))

for r in range(0, len(table[:,0]), 1):
    rt_4_y[r] = table[:, 1][r]
    rt_4_x[r] = table[:, 0][r]
    
rt_r = np.array([rt_4_x, rt_4_y])

### plot
%matplotlib notebook

plt.plot(rt_r[0], np.log(rt_r[1]))
```

```python
with open('/home/jg2364/XRD/YES187/9_28/data/th2th/9_28_002_RT', 'wb') as f:
    np.save(f, rt_r)
```
